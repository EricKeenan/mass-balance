# Combine 1980 - 2021 PRECSN, PRECLS, and PRECCU files into one PSUM file

```python
M2_PRECSN = xr.open_mfdataset("/scratch/summit/erke2265/MERRA2/PRECSN_daily_*.nc", combine='nested', concat_dim='time')
M2_PRECLS = xr.open_mfdataset("/scratch/summit/erke2265/MERRA2/PRECLS_daily_*.nc", combine='nested', concat_dim='time')
M2_PRECCU = xr.open_mfdataset("/scratch/summit/erke2265/MERRA2/PRECCU_daily_*.nc", combine='nested', concat_dim='time')
```

```python
M2_PSUM = M2_PRECSN['PRECSN'] + M2_PRECLS['PRECLS'] + M2_PRECCU['PRECCU']
M2_PSUM
```

```python
!rm M2_PSUM_1980-2021.nc
M2_PSUM.to_netcdf(path="M2_PSUM_1980-2021.nc")
M2_PRECSN.close()
M2_PRECLS.close()
M2_PRECCU.close()
```
